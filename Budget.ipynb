We replaced the irrelevent symbol from commute distance to relevant columns

Visualized the data with KPIs

```python
plt.figure(figsize=(8,4))
ax = sb.countplot(data = customers, x = 'Gender')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count of Gender')
plt.show()

customers['Gender'].value_counts()
```

As we can see males are slightly more than females in purchases

```python
plt.figure(figsize=(8,4))
ax = sb.countplot(data = customers, x = 'MaritalStatus', color= 'purple')
plt.title('Married or Unmarried Distribution')
plt.xlabel('Married/Single')
plt.ylabel('Count')
plt.show()

customers['MaritalStatus'].value_counts()
```

As we can see married persons are more than single persons in terms of purchases

```python
plt.figure(figsize=(8,4))
ax = sb.countplot(data = customers, x = 'CommuteDistance(in miles)', color= 'green')
plt.title('Commute Distance')
plt.xlabel('Commute Distance')
plt.ylabel('Count')
plt.show()
```

Maximum customers are those who are 0-1 miles away

```python
plt.figure(figsize=(8,4))
ax = sb.countplot(data = customers, x = 'YearlyIncome', color= 'blue')
plt.title('Yealy Income Distribution')
plt.xlabel('Yearly Income')
plt.ylabel('Count')
plt.xticks(rotation =90 )
plt.show()
```

There are many employees whose salary is 60,000
